```python
import json
```

```python
def getData(alamat):
    lineList = list()
    with open(alamat, encoding = "ISO-8859-1") as f:
        for line in f:
            lineList.append(line.rstrip('\n'))
    return lineList
```

```python
k = getData("last_use_k.txt")
r = getData("last_use_r.txt")
dict_ =dict()
for a, b in zip(k, r):
    dict_.update( {a : b} )
dict_
```

```python
dict_['berikan']='beri'
with open("last_use.json", 'w') as f:
    json.dump(dict_, f)
dict_
```
